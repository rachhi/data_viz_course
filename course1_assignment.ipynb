```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('sunshine.csv')
df
```

```python
alt.Chart(df).mark_line().encode(
    x='month_num',
    y='sunshine_hours',
    color='city'
)

```

```python
df_seattle = df[df['city']=='Seattle']
month_num_to_name={idx:val for idx, val in enumerate(df_seattle['month'].values)}
```

```python
alt.Chart(df).mark_line().encode(
    x=alt.X('month', title='Month', sort = alt.EncodingSortField(
                    field="month_num",order="ascending"), axis=alt.Axis(labelAngle=0)),
    y=alt.Y('sunshine_hours', title='Hours of Sunshine'),
    # color='city',sort = alt.EncodingSortField(
    #                 field="minimum(sunshine_hours)",order="ascending")
    color=alt.Color('city:N',
                    sort=['Miami', 'San Francisco', 'New York', 'Houston', 'Chicago', 'Seattle'])
)
# alt.Axis(labels=month_num_to_name.values)
```
